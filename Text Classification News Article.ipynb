```python
import pandas as pd

df = pd.read_csv('News_Category_Dataset_Updated.csv')
df.sample(5)
```

```python
df['category'].value_counts()
```

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
df.isna().sum()
```

```python
df['Text'] = df['headline'] + " " + df['short_description']
df.head()
```

# Data Cleaning

```python
import nltk

nltk.download('all')
```

```python
#lower case
df['Text'] = df['Text'].apply(lambda x: " ".join(x.lower () for x in x.split()))
```

```python
df['Text'][0]
```

```python
# removing puntuation
df['Text'] = df['Text'].str.replace('[^\w\s]', '')
```

```python
df['Text'][0]
```

```python
# removing stopwords
from nltk.corpus import stopwords 

stop_words = set(stopwords.words('english'))

df['Text'] = df['Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop_words)]))
```

```python
df['Text'][0]
```

```python
# lemmatization
from nltk.stem import WordNetLemmatizer
lemma = WordNetLemmatizer()
def lemmatize_words(text):
    words = text.split()
    words = [lemma.lemmatize(word,pos='v') for word in words]
    return ' '.join(words)
df['Text'] = df['Text'].apply(lemmatize_words)
```

```python
df['Text'][0]
```

```python
df['category'].unique()
```

```python
df['Encoded_Category'] = df['category'].astype('category').cat.codes
df.head(10)
```

```python
X = list(df['Text'])
y = list(df['Encoded_Category'])
```

```python
# dividing into training and testing set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify=list(df['Encoded_Category']))
```

```python
X_train[0]
```

```python
# Define pretrained tokenizer and model

from transformers import DistilBertTokenizerFast, TFDistilBertForSequenceClassification

tokenizer = DistilBertTokenizerFast.from_pretrained('distilbert-base-uncased')
```

```python
train_encodings = tokenizer(X_train, truncation = True, padding = True  )

val_encodings = tokenizer(X_test, truncation = True, padding = True )
```

```python
import tensorflow as tf

train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train
))

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(val_encodings),
    y_test
))
```

# Fine-tuning using native TensorFlow

```python
model = TFDistilBertForSequenceClassification.from_pretrained('distilbert-base-uncased',
                                                              num_labels=5)

loss_fn = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True)
optimizer = tf.keras.optimizers.Adam(learning_rate=1e-5, epsilon=1e-08)

#compile the model
model.compile(optimizer=optimizer,loss=loss_fn,metrics=['accuracy'])

#train the model
model.fit(train_dataset.shuffle(100).batch(16),
              epochs=1,
              batch_size=16)
```

```python
model.evaluate(test_dataset.shuffle(100).batch(16))
```

```python
len(val_encodings)
```

```python
len(tf.data.Dataset.from_tensor_slices(
    dict(val_encodings)))
```

```python
y_pred = []
def get_pred(text):

    output = model.predict(text)[0]

    prediction_value = tf.argmax(output, axis = 1).numpy()[0]

    y_pred.append(prediction_value)
    
for word in tf.data.Dataset.from_tensor_slices(
    dict(val_encodings)):
    get_pred(word)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_pred, y_test))
```

# Saving & Loading the model

```python
saved_model = "/text_classification_model" 

model.save_pretrained(saved_model)

tokenizer.save_pretrained(saved_model)
```
